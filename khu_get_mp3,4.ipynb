```python
!pip install moviepy requests

import re
import requests
from moviepy.editor import VideoFileClip

def parse_curl(curl_command: str):
    """
    주어진 cURL 명령문에서 URL과 HTTP 헤더들을 추출합니다.

    Args:
        curl_command (str): 여러 줄로 구성된 cURL 명령문

    Returns:
        tuple: (URL 문자열 또는 None, 헤더들을 담은 dict)
    """
    # Windows 스타일의 ^ 기호와 줄바꿈 제거
    cleaned = curl_command.replace('^', '').replace('\n', ' ')

    # URL 추출: curl 다음에 나오는 큰따옴표(") 안의 문자열 찾기
    url_match = re.search(r'curl\s+"([^"]+)"', cleaned)
    url = url_match.group(1) if url_match else None

    # 모든 -H "Header: value" 부분 추출
    header_matches = re.findall(r'-H\s+"([^"]+)"', cleaned)
    headers = {}
    for header in header_matches:
        parts = header.split(":", 1)
        if len(parts) == 2:
            key = parts[0].strip()
            value = parts[1].strip()
            headers[key] = value
    return url, headers
```

```python
for h in ("Range", "range", "If-Range", "if-range"):
    headers.pop(h, None)


# --- 영상 다운로드 ---
video_filename = "screen.mp4"
audio_filename = "screen_audio.mp3"

print("\n영상 다운로드 중...")
response = requests.get(url, headers=headers, stream=True)
response.raise_for_status()  # 오류 발생 시 예외 처리

with open(video_filename, "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)
print("영상 다운로드 완료:", video_filename)

# --- 오디오 추출 ---
print("오디오 추출 중...")
video_clip = VideoFileClip(video_filename)
audio_clip = video_clip.audio
audio_clip.write_audiofile(audio_filename)
print("오디오 추출 완료:", audio_filename)
```
